```python
import csv
import matplotlib.pyplot as plt

f= open('2023년 03월  교통카드 통계자료.csv',encoding='utf-8')
data = csv.reader(f)

mx=0
mx_station=[]
result_ride=[]
result_stop=[]
result=[]

for row in data:
    mx_station.append(row[3])
    row[4:]=map(int,row[4:])
    result_ride.append(sum(row[10:15:2]))
    result_stop.append(sum(row[11:15:2]))
    
for i in range(len(result_ride)):
    result.append(result_ride[i] + result_stop[i])
#
top_stations_ride = [x for _, x in sorted(zip(result_ride, mx_station), reverse=True)][:30]
top_result_ride=sorted(result_ride, reverse=True)[:30]

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('출근시간 최대 승차역 30개')
plt.bar(range(30),top_result_ride, color='b')
plt.xticks(range(30),top_stations_ride,rotation=90)
plt.show()

print("출근 시간(7 ~ 9시), 사람들이 가장 많이 승차하는 역은",top_stations_ride[0],"역입니다.")
print("해당 역 승차객 수: ",top_result_ride[0],"명")
print()

#
top_stations_stop = [x for _, x in sorted(zip(result_stop, mx_station), reverse=True)][:30]
top_result_stop=sorted(result_stop, reverse=True)[:30]

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('출근시간 최대 하차역 30개')
plt.bar(range(30),top_result_stop, color='b')
plt.xticks(range(30),top_stations_stop,rotation=90)
plt.show()

print("출근 시간(7 ~ 9시), 사람들이 가장 많이 하차하는 역은",top_stations_stop[0],"역입니다.")
print("해당 역 하차객 수: ",top_result_stop[0],"명")
print()

#
top_stations = [x for _, x in sorted(zip(result, mx_station), reverse=True)][:30]
top_result=sorted(result, reverse=True)[:30]
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('출근시간 최대 승하차역 30개')
plt.bar(range(30),top_result, color='b')
plt.xticks(range(30),top_stations,rotation=90)
plt.show()

print("출근 시간(7 ~ 9시), 사람들이 가장 많이 승하차하는 역은",top_stations[0],"역입니다.")
print("해당 역 승하차객 수: ",top_result[0],"명")
print()

#
print("따라서, 출근 시간대 사람들이 가장 많이 타고 내리는 역은",top_stations[0],"역입니다.")
```
