```python
import matplotlib.pyplot as plt
import random
import csv

dice_a =[]
dice_b =[]
dice_c =[]
dice_d =[]

#
for i in range(100):
    dice_a.append(random.randint(1,6))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('주사위 100번 시행 시뮬레이션1 결과')
plt.hist(dice_a, bins=6)
plt.xlabel("Rolling dice", size=12)
plt.ylabel("Frequency_Histogram", size=12)
plt.show()

#
for i in range(1000):
    dice_b.append(random.randint(1,6))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('주사위 1000번 시행 시뮬레이션2 결과')
plt.hist(dice_b, bins=6)
plt.xlabel("Rolling dice", size=12)
plt.ylabel("Frequency_Histogram", size=12)
plt.show()

#
for i in range(10000):
    dice_c.append(random.randint(1,6))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('주사위 10000번 시행 시뮬레이션3 결과')
plt.hist(dice_c, bins=6)
plt.xlabel("Rolling dice", size=12)
plt.ylabel("Frequency_Histogram", size=12)
plt.show()

#
for i in range(100000):
    dice_d.append(random.randint(1,6))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('주사위 100000번 시행 시뮬레이션4 결과')
plt.hist(dice_d, bins=6)
plt.xlabel("Rolling dice", size=12)
plt.ylabel("Frequency_Histogram", size=12)
plt.show()

#
results = [dice_a, dice_b, dice_c, dice_d]
with open('dice_simulation_results.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Simulation', 'Result'])
    for i, result in enumerate(results):
        for value in result:
            writer.writerow([f'Simulation {i+1}', value])

#
print(" 주사위를 던지는 횟수가 증가할수록, 각 숫자가 나올 확률이 균등해진다는 것을 알 수 있습니다. 위 그래프들을 비교해보면, 주사위를 던지는 횟수가 적을 때는 특정한 숫자가 상대적으로 많이 나오는 경향이 있지만, 시행 횟수가 증가할수록 숫자가 나올 확률이 비슷해지는 것을 확인할 수 있습니다.")
print(" 시행 횟수가 많아질수록, 주사위를 던질 때 얻을 수 있는 숫자들의 분포가 정규분포에 가까워진다는 것을 알 수 있습니다. ")
```
