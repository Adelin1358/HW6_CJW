```python
import csv 
import matplotlib.pyplot as plt

f= open('201312_202212_주민등록인구및세대현황_연간(1).csv','r',encoding='ANSI')
data = csv.reader(f)
header =next(data)

m=[]
f=[]
p=[]

for row in data:
    for i in row[1::3]:
        m.append(int(i))
    for i in row[2::3]:
        f.append(int(i))
    for i in row[3::3]:
        p.append(float(i))
        
for i in range(0,10):
    plt.rc('font', family='Malgun Gothic')
    plt.rcParams['axes.unicode_minus']=False
    year=i+2013
    plt.title(str(year)+'년도 제주특별자치도 남녀 성별 비율')
    wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
    plt.axis('equal')
    plt.pie([m[i],f[i]],labels=['남','여'],autopct='%.1f%%',startangle=90, wedgeprops= wedgeprops)
    plt.show()

    print(year,"년도 제주특별자치도 남자 인구수: "+ str(m[i])+" 명")
    print(year,"년도 제주특별자치도 여자 인구수: "+ str(f[i])+" 명")
    print("남여 비율: ",p[i])
    
    if (m[i]>f[i]):
        print(year,"년도 제주특별자치도의 남녀 인구수 분석 결과, 남자가 여자보다",m[i]-f[i],"명 더 많습니다.\n")
    elif (m[i]<f[i]):
        print(year,"년도 제주특별자치도의 남녀 인구수 분석 결과, 여자가 남자보다 ",f[i]-m[i],"명 더 많습니다.\n")
    else:
        print(year,"년도 제주특별자치도의 남녀 인구수 분석 결과, 남,녀 인구수가 서로 같습니다.\n")

print("<결론>")        
print(" 2013~2022년도 제주특별자치도의 남녀 인구수를 비교한 결과, 2013~2022년도 사이에는 남자 인구수가 여자 인구수보다 더 많은 것으로 집계되었습니다.")
print(" 따라서, \'제주도에는 여성의 비율이 더 높다\'는 속설은 틀렸습니다.")
```
